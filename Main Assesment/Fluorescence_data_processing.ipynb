```python
import pandas as pd
import numpy as np

blank_rows = np.arange(1,12,1)
fluor_conc = []
for i in range(0,11):
    fluor_conc.append(10/(2**i))
row_data = {
    'Row' : blank_rows,
    'Fluor_conc' : fluor_conc
}
blank_corrected = pd.DataFrame(row_data)

print(blank_corrected)

```

```python
dfs = {}
for i in range(1,13):
    dfs[f"df{i}_bc"] = pd.read_excel(io = 'Main Assesment/Data/pd'+str(i)+'.xlsx',sheet_name = 'End Point', header = 0, range = 'B15:L22')
    avg = []
    for col in range(1,12):
        avg.append(
            np.average(dfs[f"df{i}_bc"].iloc[col])
        )
    dfs[f"df{i}_bc"].insert(avg)
```
