```python
import pandas as pd
import numpy as np

pb_exp_num = np.arange(1,13,1)

df_900 = {}
df_975 = {}
avg_900 = []
avg_975 = []
std_900 = []
std_975 = []

for i in range(1,13):
    df_900[f"df{i}"] = pd.read_excel(
        io = '/Users/wilsonporteus/Documents/GitHub/ssb_s5/Main Assesment/Data/pb'+str(i)+'a.xlsx',
        header=0,
        sheet_name = 'End point', usecols="B:L",
        skiprows=13,
        nrows=9, engine="openpyxl"
        )
    df_975[f"df{i}"] = pd.read_excel(
        io = '/Users/wilsonporteus/Documents/GitHub/ssb_s5/Main Assesment/Data/pb'+str(i)+'a.xlsx',
        header=0,
        sheet_name = 'End point', usecols="B:L",
        skiprows=24,
        nrows=9, engine="openpyxl"
        )
    a = np.average(df_900[f"df{i}"])
    avg_900.append(a)
    b = np.average(df_975[f"df{i}"])
    avg_975.append(b)
    c = np.std(df_900[f"df{i}"].to_numpy().ravel(), ddof=1)
    std_900.append(c)
    d = np.std(df_975[f"df{i}"].to_numpy().ravel(), ddof=1)
    std_975.append(d)
    

pb_data = {
    'Experiment Number' : pb_exp_num,
    '900 Avg' : avg_900,
    '900 Std' : std_900,
    '975 Avg' : avg_975,
    '975 Std' : std_975
}
pb_df = pd.DataFrame(pb_data)

pb_df['Avg'] = pb_df['975 Avg']-pb_df['900 Avg']
pb_df['Std'] = pb_df['975 Std']-pb_df['900 Std']
pb_df['CV'] = pb_df['Std']/pb_df['Avg']
pb_df['CV'].to_csv('PB_CV.csv',index=False)
```
